Question 2 asks to find which factors can be used to distinguish job titles from each other. I decided to tackle this as a classification problem as in this case the target is a categorical value.

```python
import pandas as pd
import numpy as np
from itertools import cycle
import matplotlib.pyplot as plt
import re
import scipy as sp
import seaborn as sns
from sklearn.model_selection import cross_val_score, train_test_split, cross_val_predict
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression, LinearRegression, LassoCV, Lasso
from sklearn import metrics,svm
from textblob import TextBlob, Word
from nltk.stem.snowball import SnowballStemmer
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,RandomForestClassifier
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
from nltk.stem import LancasterStemmer, WordNetLemmatizer,PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from sklearn.preprocessing import label_binarize
from sklearn.multiclass import OneVsRestClassifier
from scipy import interp
import pydotplus
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
from xgboost import XGBClassifier
```

```python
#Loading the correct csv into pandas
jobs=pd.read_csv('model_ready.csv',index_col=0)
```

## Choosing the dependant and independant variables

### Independant variables
The independant variables I will be using to make my models will be the descriptions, states and cities. I did not chose to include relative titles or categories because I used that information to compute what the salaries should be so including them would artificailly improve my model due to their high correlation.

#### Coding dummy variables
I used pandas to encode dummy variables for the cities and states for the jobs so I can use them for moddeling

```python
df_dummies = pd.get_dummies(jobs[['city','state']], drop_first=True)
df_dummies.shape
```

```python
#I created a new dataframe concatenating the independant variables I have chosen
X_df = pd.concat([df_dummies, jobs.cleaned_descriptions], axis=1)
```

### Dependant variable
For this question I decided to try to predict the job title for each of the jobs. These were titles that were provided in the data when I scrapped the job website and I think I have a good amount of information about different job titles within the description to make accurate predictions.

```python
y=jobs.rel_title
```

#### Baseline prediction
The baseline prediction for this problem is 67.6% because this is the ratio for the most predominant group in my dataset.

Although this baseline would not be too much of an issue if it was simply a binary class, because there are three classes, there is quite an imbalance between the three classes. This may need to be addressed at a later time but for now I will leve it as is.

```python
y.value_counts() / len(y)
```

### Splitting to train and test data

Here I split my data into a train and test set. I will be training all my models on the training set and testing on the test set. This will give me the best idea of how well generalized my model is. I opted with a training set size of 70%.

```python
X_train, X_test, y_train, y_test = train_test_split(X_df, y, random_state=1,train_size=0.7,stratify=y)
```

## Vectorization
Because we want to identify specifically what determined a job title, I thought it was important to document some the different vectorizers that can be used on the data and how they might affect the interpretability of our model.

### Count vectorizer
This convert a collection of text documents to a matrix of token counts which makes it a format understandable by a model.

```python
#Implimented a normal count cevtorizer with no stop words or any parrams
vect = CountVectorizer(max_features=1000,ngram_range=(1,3))
X_train_dtm = vect.fit_transform(X_train.cleaned_descriptions)
X_test_dtm = vect.transform(X_test.cleaned_descriptions)
```

```python
#Made it a dataframe
train_count_vect_df = pd.DataFrame(X_train_dtm.todense(), columns=vect.get_feature_names())
test_count_vect_df = pd.DataFrame(X_test_dtm.todense(), columns=vect.get_feature_names())
```

```python
#Created a compiled dataframe with the other independant variables added
X_train_dtm=pd.concat([X_train.drop('cleaned_descriptions',axis=1).reset_index(drop=True), train_count_vect_df], axis=1)
X_test_dtm=pd.concat([X_test.drop('cleaned_descriptions',axis=1).reset_index(drop=True), test_count_vect_df], axis=1)
```

### Count vectorizer - Logistic regression
Because this is a classification problem I have fit a logistic regression to begin with but with no specific parramaters used  to get a rough idea of how the model might perform.

```python
logr = LogisticRegression()

logr.fit(X_train_dtm, y_train);
```

#### Model evaluation
Right off the bat, the logistic regression model is performing quite well. I am returning an average f1 score of 93 on the test data which is better than the baseline.

Interestingly, most of my predictions fail around the title of Data Analyst. Where the scores for Business Analyst and Data Scientist are above **0.95**, Data Analyst scores or in the high **70s**.

```python
#Retrieve predicted y values to use in scoring
y_pred=logr.predict(X_test_dtm)
logr_score=metrics.classification_report(y_test, y_pred)
print(logr_score)
```

### L1 penalty - logistic regression
I used the L1 penalty jsut like I did with the regression problem to try to reduce the amount of variables being considered by the model and to improve computational time.

```python
#Apply penalty and refit the model
logr_v1 = LogisticRegression(penalty='l1', solver='saga')

logr_v1.fit(X_train_dtm, y_train)
logr_v1_cross = cross_val_score(logr_v1, X_train_dtm, y_train, cv=4)
print('Mean L1 penatly logistic regression CV score:',np.mean(logr_v1_cross));
```

#### Model evaluation
Again this prodcued quite a decent result, interestingly slightly worse than the regular logistic regression in terms of the average scores. Whether or not this is worth the tradeoff will depend on how the model might be implemented. One of the main questions that will determine ifthe tradeoff is worth it is, will we need to perform this prediction real time? If there is a need to, a **0.03** drop of f1score may be worth the reduced dimensionality and improved

```python
y_pred=logr_v1.predict(X_test_dtm)
logr_score_v1=metrics.classification_report(y_test, y_pred)
print(logr_score_v1)
```

#### Model interpretation
##### Exploring the model
Using the attributes of my model, I retreived all the logg odds and lined them up with the appropriate classes.

```python
#Creating a dataframe matching logg odds with classes
logr_coefs = pd.DataFrame(columns=X_train_dtm.columns,data=logr_v1.coef_,index=logr_v1.classes_)
```

```python
#Because I am using Lasso penalty, there will be some coefs with a value of 0, this will remove them
logr_coefs=logr_coefs.T.loc[(logr_coefs.apply(sum)>0)].T
```

##### What are the key factors?
From the output of the model, because the coeficients printed below suggest that the weightage is actually on words that give no real meaning, this means I should move to stop words and TFIDF instead of just Count vectorizer.

```python
#Printing the top 8 values in terms of coef for each class 
for i,e in enumerate(logr.classes_):
    print(e)
    print(np.abs(logr_coefs.iloc[i,:]).sort_values(ascending=False).head(8))
    print('------------------------------------\n')
```

### Stop words

Because there are still currently irrelevant words within my description, I will be using and implementing stopwords from the NLTK library for english. this is basically a list of commonly used words in text that don't normally provide any meaningfull information. To that list of words, I appended the titles that I will be classifying for because if the description has those words it would be pointless in trying to predict the title with them.

```python
#Call relevant stopwords
sw=stopwords.words('english')
```

I decided to add some extra stop words to the list as I notice these were popping up in the sorted log odds list printed above.

```python
#Create a list of words to add to default stopwords
key_words=['data','analyst','scientist','business','senior','junior','use','using','com','au']
for i in key_words:
    sw.append(i)
```

#### Apply stop words to vectorizer
I have kept the other variables the same and added the stop words to the model.

```python
vect = CountVectorizer(ngram_range=(1,3),stop_words=sw,max_features=1000)
X_train_dtm = vect.fit_transform(X_train.cleaned_descriptions)
X_test_dtm = vect.transform(X_test.cleaned_descriptions)
```

```python
train_count_vect_df = pd.DataFrame(X_train_dtm.todense(), columns=vect.get_feature_names())
test_count_vect_df = pd.DataFrame(X_test_dtm.todense(), columns=vect.get_feature_names())
```

```python
X_train_dtm=pd.concat([X_train.drop('cleaned_descriptions',axis=1).reset_index(drop=True), train_count_vect_df], axis=1)
X_test_dtm=pd.concat([X_test.drop('cleaned_descriptions',axis=1).reset_index(drop=True), test_count_vect_df], axis=1)
```

```python
logr_v2 = LogisticRegression(penalty='l1', solver='saga')

logr_v2.fit(X_train_dtm, y_train);
```

#### Model evaluation
Using this model removing the stopwords, the score overall has dropped but the information gained is much more relevant.

```python
y_pred=logr_v2.predict(X_test_dtm)
logr_score_v2=metrics.classification_report(y_test, y_pred)
print(logr_score_v2)
```

#### Model interpretation
There are some interesting variables within each classifier. For Business alayst, banking is a good deterimant, sql and exel are important for data analysts and python and machine learning are important for data scientists.

The problem with this model is I would like some more in depth information, more precisely to do with what skills are relevant and I believe moving from count vectorizer to tfidf will help in that regard as wells as using n-grams.

```python
logr_coefs = pd.DataFrame(columns=X_train_dtm.columns,data=logr_v2.coef_,index=logr_v2.classes_)
logr_coefs=logr_coefs.T.loc[(logr_coefs.apply(sum)>0)].T
for i,e in enumerate(logr_v2.classes_):
    print(e)
    print(np.abs(logr_coefs.iloc[i,:]).sort_values(ascending=False).head(8))
    print('------------------------------------\n')
```

## TFIDF - Logistic regression
I will now explore the affect of using TFIDF which will give weightage on words which have more frequency within a text. I anticipate with will improve the interpretability of the results and also allow me to make clearer reconmendations. I have also decided to apply n-grams to again improve my ability to understand exactly what is important for each of the classes.

```python
vect = TfidfVectorizer(ngram_range=(2,3),stop_words=sw,max_features=1000)
X_train_dtm1 = vect.fit_transform(X_train.cleaned_descriptions)
X_test_dtm1 = vect.transform(X_test.cleaned_descriptions)
```

```python
train_count_vect_df = pd.DataFrame(X_train_dtm1.todense(), columns=vect.get_feature_names())
test_count_vect_df = pd.DataFrame(X_test_dtm1.todense(), columns=vect.get_feature_names())
```

```python
X_train_dtm1=pd.concat([X_train.drop('cleaned_descriptions',axis=1).reset_index(drop=True), train_count_vect_df], axis=1)
X_test_dtm1=pd.concat([X_test.drop('cleaned_descriptions',axis=1).reset_index(drop=True), test_count_vect_df], axis=1)
```

```python
logr_v3 = LogisticRegression(penalty='l1', solver='saga')

logr_v3.fit(X_train_dtm1, y_train);
```

### Model evaluation
Although I expected incorporating TFIDF would improve the results, this however isn't the case. In fact the model actually performed much worse than anticipated when looking at all the averaged scores and especially in reference to Data analyst.

```python
y_pred=logr_v3.predict(X_test_dtm1)
logr_score_v3=metrics.classification_report(y_test, y_pred)
print(logr_score_v3)
```

```python
mat = metrics.confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8,8))
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False,cmap='GnBu')
plt.xticks([0.5,1.5,2.5],logr_v3.classes_,rotation=0)
plt.yticks([0.5,1.5,2.5],logr_v3.classes_,rotation=360)
plt.xlabel('True label')
plt.ylabel('Predicted label')
plt.title('Confusion matrix heatmap');
```

### Model interpretation
You can see from the coefficients that there is much more actionable results within the coefficients. Instead of single word variables, the bi grams give alot of information regarding what is important for each class. It is important to note that the coef I have returned are the abssolute values jsut to show the highest impact variables.
