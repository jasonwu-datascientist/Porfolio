# Project overview
---
This project is focused on exploratory data analysis, aka "EDA". EDA is an essential part of the data science analysis pipeline. Failure to perform EDA before modeling is almost guaranteed to lead to bad models and faulty conclusions. My goal is to lay down the layout for what are good practices for all projects going forawrd and demonstrate the importance of EDA.

## Load the dataset and describing it
---
I had to load the csv using the reader function in python and separated the header from the data. I then used a function that takes the header information and data information to make a new dictionary.

```python
#Assign an object with the relative path
sat_scores = '../sat_scores.csv'

with open(sat_scores) as csvfile:
    reader=csv.reader(csvfile, delimiter=',')
    csvlist=[]
    for row in reader:
        csvlist.append(row)
```

```python
#Assign the header from the data
header=csvlist[0]
data=csvlist[1:]
```

```python
#A function to convert the data into a dictionary format
def converttodict(headerlist,datalist):

    newdict={}

    for index,i in enumerate(header):
        newdict.update({i:[]})

        for k in data:
            
            
            newdict[i].append(k[index])
            
    return(newdict)
```

```python
#Running the function and assigning it to an object called Dictionary1
Dictionary1=converttodict(header,data)
```

```python
drug_use=pd.read_csv('../drug-use-by-age.csv', index_col='age')
```

#### Interactive plot using ipywidgets
---
To make it easier to visualise on a larger plot I created an interactive graph.

```python
def power_plotter(x):

    fig, axes = plt.subplots(1 ,figsize=(30,6))
    
    drug_use[x].plot()
    plt.title(x,fontsize=30)
    plt.xlabel('Age', fontsize=18)
    plt.ylabel('Percentage use',fontsize=18)
    plt.yticks(fontsize=20)
    plt.xticks(np.arange(17), indexes, fontsize=20)
    plt.show()

interact(power_plotter, x=drug_use_percent);
```

The basis of my statistical tests was to basically see if there were any statistically significant differences in the means of the drug groups. In all of my cases this meant that -

H0 = The means for drug percentage are not statistically different

H1 = The means for drug percentage are statistically different

Due to our pvalue<0.01, we reject the H0 in this case


```python
newdf=drug_use[drug_use_percent].iloc[:,1:]
drug_use['totaluse']=newdf.T.sum()
```

I decided to explore the relationship between alcohol and total drugs other than alcohol. I thought it would be a good idea to use more hypothesis testing to to explore my previous results.

In this case I used Manwhiteney U method-

H0 = The means between total use other than alcohol and alcohol use are statistically similar

H1 = The means between total use other than alcohol and alcohol use are not statistically similar


Due to the output of this test being P>0.01, this means that we accept the H0

# Linear regression
---
I decided to fit a very rudimentary linear regression model onto the data to observe how well a model might do in predicting drug usage.

```python
#Defining a function to do linear regression from scratch
def linerreg(x,y,df):
    
    #Assigning the first coefficient
    b1=(stats.pearsonr(x,y)[0])*np.std(y)/np.std(x)
    
    #Assigning the intercept
    b0=np.mean(y)-(b1*np.mean(x))
    
    #Collecting predictions using the model
    y_pred=[]
    for i in x:
        b = b0 + (b1*i)
        y_pred.append(b)
    
    #Creating a new dataframe that will be used to plot the simple linear regression model
    df['prediction']=y_pred
    print('Optimal slope',b1)
    print('Optimal intercept', b0)
    fig=plt.figure(figsize=(10,10))
    plt.scatter(x,y)
    plt.plot(x, df['prediction'], color='r')
    plt.xlabel(x.name, fontsize=20)
    plt.ylabel(y.name, fontsize=20)
    plt.yticks(fontsize=15)
    plt.xticks(fontsize=15)
    plt.title("Linear regression", fontsize=30)
    yhat=b0 + b1*x
    
    #Calculating R^2 value
    err_tot = (y-np.mean(y))**2
    err_res = (y - yhat)**2
    ss_tot = np.sum(err_tot)
    ss_res = np.sum(err_res)
    print('Regression R^2:', 1 - (ss_res/ss_tot))
    plt.savefig('SLR.png')
```

```python
linerreg(drug_use.alcohol_use,drug_use.totaluse,drug_use)
```
